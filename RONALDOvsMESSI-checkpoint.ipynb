# CRISTIANO RONALDO VS LIONEL MESSI 

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as ply
import plotly.express as px
import plotly.graph_objects as go
```

```python
df=pd.read_csv('cristiano_vs_messi.csv')
```

### DATA MANIPULATION


```python
df['comp']=df['comp'].ffill()
df['date']=df['date'].ffill()
df['round']=df['round'].ffill()
df['venue']=df['venue'].ffill()
df['opp']=df['opp'].ffill()
```

```python
df['min']=df['min'].apply(lambda x:x.replace("'",''))
df['min']=df['min'].apply(lambda x:x.replace("+",''))
```

```python
df.tail()
```

```python
df['min']=pd.to_numeric(df['min'])
df['time_class']=df['min'].apply(lambda x:'first_half' if x<=45 else ( 'secound_half' if 45<x<=90 else 'extra_time'))
df['assist']=df['assisted'].fillna(0)
df['solo']=df['assist'].apply(lambda x:'solo' if x==0 else 'assisted')
df.head()
```

```python
from datetime import date
df['date']=pd.to_datetime(df['date'])

L = ['year', 'month', 'day', 'dayofweek', 'dayofyear', 'weekofyear', 'quarter']
df = df.join(pd.concat((getattr(df['date'].dt, i).rename(i) for i in L), axis=1))
df['dayofweek']=pd.to_numeric(df['dayofweek'])
df['dayofweek']=df['dayofweek'].apply(lambda x:x+1)
```

```python
df.tail()
```

## NUMBER OF GOALS

```python
df['player'].value_counts().plot(colormap='jet',kind='line')
```

```python
df['player'].value_counts().plot(colormap='plasma',kind='barh')
```

## GOALS SCORED - ASSISTED VS SOLO

```python
df_ronaldo=df.loc[df['player']=='ronaldo']
df_messi=df.loc[df['player']=='messi']

ronaldo_solo=df_ronaldo[df_ronaldo['solo']=='solo'] 
ronaldo_assisted=df_ronaldo[df_ronaldo['solo']=='assisted'] 
slices=[len(ronaldo_solo),len(ronaldo_assisted)]
labels=['wihtout_assist','assisted']
 
messi_solo=df_messi[df_messi['solo']=='solo'] 
messi_assisted=df_messi[df_messi['solo']=='assisted'] 
slices1=[len(messi_solo),len(messi_assisted)]
labels1=['without_assist','assisted']

fig, axes = ply.subplots(1, 2, figsize=(15, 5), sharey=False)
fig.suptitle('Cristiano Vs Messi/  without-assist & assisted')
axes[0].pie(slices,labels=labels,startangle=45,shadow=2,explode=(0,0.1),autopct='%1.2f%%',colors=['orange','yellow']);
axes[0].set_title('Cristiano Ronaldo')
axes[1].pie(slices1,labels=labels1,startangle=45,shadow=2,explode=(0,0.1),autopct='%1.2f%%',colors=['cyan','blue']);
axes[1].set_title('Lionel Messi');
```

## GOALS SCORED ( 1st half, 2nd half, Extra time)

```python
ply.figure(figsize=(15,6))
ply.suptitle('When do they score the goals')

ply.subplot(1,3,1)
sns.countplot(data=df[df.time_class=='first_half'],x='player',palette='deep').set_title('1st Half')

ply.subplot(1,3,2)
sns.countplot(data=df[df.time_class=='secound_half'],x='player',palette='deep').set_title('2nd Half')

ply.subplot(1,3,3)
sns.countplot(data=df[df.time_class=='extra_time'],x='player',palette='deep').set_title(' Extra Time')

ply.show()
```

## MOST ASSISTS RECEIVED FROM:
